```python
#solving decay equation by Euler's first order method
import numpy as np
import matplotlib.pyplot as plt

#initial conditions
N_0 = 100       #هسته اولیه
T=1             #ثابت زمانی
delta_t=5       #total time
h=0.05          #گام زمانی

#تعداد گام های زمانی
m=int(delta_t/h)
    
#لیست های زمان و هسته برای روش اویلر
t_list = [0]
N_list = [N_0]

#الگوریتم اویلر#
for n in range(1,m):
        t_new=n*h
        N_new=N_list[-1]+(-h/T)*(N_list[-1])
        t_list.append(t_new)
        N_list.append(N_new)

#تابع تحلیلی
t_analytical = np.linspace(0,delta_t,500)
N_analytical = (N_0)*np.exp(-t_analytical/T)

#تبدیل لیست ها به آرایه
t_euler = np.array(t_list)
N_euler = np.array(N_list)

#رسم نمودار
plt.plot(t_euler, N_euler, color ='red', label= 'Numerical')
plt.plot(t_analytical, N_analytical,'*', color = 'yellow', label= 'Analytic')
plt.xlabel('Time')
plt.ylabel('N(t)')
plt.legend()
plt.show()
```
